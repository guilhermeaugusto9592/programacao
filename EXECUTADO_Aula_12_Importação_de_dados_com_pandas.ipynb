### Extraindo consultas de um banco de dados.

```python
import sqlite3
import pandas as pd
```

```python
conec_banco = sqlite3.connect('/content/drive/MyDrive/Programação em Python para Data Science/Exemplo 6 -  loja_de_roupas.db')
```

```python
dados_produtos = pd.read_sql_query('''

SELECT * FROM produtos;

''', conec_banco)
```

```python
dados_produtos.head()
```

Para saber o nome das tabelas do banco.

```python
dados_tabelas = pd.read_sql_query('''

SELECT name FROM sqlite_master
WHERE type='table';

''', conec_banco)

dados_tabelas
```

```python
dados_colunas = pd.read_sql_query('''

SELECT * FROM lojas;

''', conec_banco)

print(dados_colunas.columns.tolist())
```
